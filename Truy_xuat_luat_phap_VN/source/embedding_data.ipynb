```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim
from pyvi import ViTokenizer
import torch
import pandas as pd
from tqdm.auto import tqdm
```

```python
MODEL_NAME = "bqbbao6/vietnamese-legal-embedding-wsgm"

# Load data
df = pd.read_csv("/content/drive/MyDrive/IE103/Project/legal_chunks.csv")

# Load the model with remote code trust
model = SentenceTransformer(MODEL_NAME, trust_remote_code=True)

# Helper function for legal formatting with word segmentation
def process_text(text, prefix="passage: "):
    if not isinstance(text, str):
        text = str(text)
    # Segment words (e.g., "Người lao động" -> "Người lao_động")
    segmented_text = ViTokenizer.tokenize(text)
    return prefix + segmented_text

print("Processing and encoding embeddings...")

# Apply word segmentation and prefixing to all texts with tqdm progress bar
texts_to_embed = [process_text(t) for t in tqdm(df["embed_text"].tolist(), desc="Preprocessing texts")]

# Generate embeddings with built-in progress bar enabled
embeddings = model.encode(
    texts_to_embed,
    batch_size=256,
    show_progress_bar=True,
    normalize_embeddings=True
)

# Create result DataFrame
df_embedding = pd.DataFrame({
    "node_id": df["node_id"],
    "embedding": [emb.tolist() for emb in embeddings]
})

# Save to CSV
df_embedding.to_csv(
    "legal_chunk_embeddings.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Done!")
print(df_embedding.head())
```
