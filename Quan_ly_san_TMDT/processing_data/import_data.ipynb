```python
import os

import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
from sqlalchemy.engine import URL
import psycopg2

load_dotenv()
```

```python
# 1. Thông tin kết nối lấy từ file .env
USERNAME = os.getenv('DB_USERNAME')
PASSWORD = os.getenv('DB_PASSWORD')
HOST = os.getenv('DB_HOST')
PORT = os.getenv('DB_PORT', '5432')
DB_NAME = os.getenv('DB_NAME', 'postgres')

required_values = {
    'DB_USERNAME': USERNAME,
    'DB_PASSWORD': PASSWORD,
    'DB_HOST': HOST,
}
missing_values = [name for name, value in required_values.items() if not value]
if missing_values:
    raise ValueError(f"Thiếu biến môi trường: {', '.join(missing_values)}")

# 2. Tạo Connection String an toàn
connection_url = URL.create(
    'postgresql+psycopg2',
    username=USERNAME,
    password=PASSWORD,
    host=HOST,
    port=int(PORT),
    database=DB_NAME,
    query={'sslmode': 'require'},
)

# 3. Tạo engine kết nối
engine = create_engine(connection_url)
```

```python
def upload_csv_to_postgres(file_path, table_name):
    try:
        df = pd.read_csv(file_path)
        

        df.to_sql(table_name, engine, if_exists='append', index=False, method='multi')
        
        print(f"Thành công: Đã nạp dữ liệu từ {file_path} vào bảng {table_name}")
    except Exception as e:
        print(f"Lỗi khi xử lý {table_name}: {e}")
```

```python
upload_csv_to_postgres('Project\data\products_dataset.parquet', 'products')
```
