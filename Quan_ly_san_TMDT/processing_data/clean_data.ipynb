```python
import pandas as pd
```

# Seller, Customer, Geolocation

```python
geolocation = pd.read_csv('Project\data\silver\geolocation_new.csv')
```

```python
geolocation
```

```python
valid_zip_codes = geolocation['geolocation_zip_code_prefix']
```

```python
customers = pd.read_parquet(r'Project\data\raw\olist_customers_dataset.parquet')
```

```python
customers = customers[customers['customer_zip_code_prefix'].isin(valid_zip_codes)]
```

```python
customers.to_csv('Project\data\silver\customers_clean.csv',index=False)
```

```python
sellers = pd.read_parquet(r'Project\data\raw\sellers.parquet')
```

```python
sellers = sellers[sellers['seller_zip_code_prefix'].isin(valid_zip_codes)]
```

```python
sellers.to_csv('Project\data\silver\sellers_clean.csv',index = False)
```

# Oders

```python
customers = pd.read_csv('Project\data\silver\customers_clean.csv')
```

```python
orders = pd.read_parquet(r'Project\data\raw\orders_dataset.parquet')
```

```python
orders = orders[orders['customer_id'].isin(customers['customer_id'])]
```

```python
orders.to_csv('Project\data\silver\orders_clean.csv',index = False, na_rep='')
```
